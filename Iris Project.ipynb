# Supervised ML : Simple Linear Regression

```python
x = 50 * np.random.rand(100)
```

```python
y = -3 * x + np.random.randn(100)
```

```python
plt.scatter(x,y);
```

# Step 1: Choose a class of model


```python
from sklearn.linear_model import LinearRegression
```

# Step 2: Select the model hyperparameters

```python
model = LinearRegression(fit_intercept=True)
```

```python
model
```

# Step 3: Arrange data into feature matrix and target Array

```python
X = x.reshape(-1,1)
X.shape
```

# Step 4: Fit model to the data

```python
model.fit(X, y)
```

```python
model.coef_
```

```python
model.intercept_
```

# Step 5: Apply trained model to new data


```python
x_fit = np.linspace(-1, 11)
```

```python
X_fit = x_fit.reshape(-1,1)
```

```python
y_fit = model.predict(X_fit)
```

# Visualize

```python
plt.figure(figsize =(10,8))
plt.scatter(x,y)
plt.plot(x_fit ,y_fit)
```
